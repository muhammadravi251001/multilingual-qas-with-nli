```python
file_name = "dev-context-en-question-en.json"
```

```python
import pandas as pd

# Membaca berkas JSON dengan orient='records' dan lines=True
with open(file_name, 'r', encoding='utf-8') as file:
    data = [line for line in file]

# Gabungkan baris-baris JSON menjadi satu string
json_data = ''.join(data)

# Baca JSON dan konversi ke DataFrame
df = pd.read_json(json_data, orient='records', lines=True, encoding='utf-8')

# Tampilkan DataFrame
df
```

```python
data_qas_train = df.loc[0, "data"]
```

```python
for i in range(len(data_qas_train)):
    for j in range(len(data_qas_train[i]['paragraphs'])):
        for k in range(len(data_qas_train[i]['paragraphs'][j]['qas'])):
        
            content = len(data_qas_train[i]['paragraphs'][j]['qas'][k]['answers'])

            print(f"i: {i}, j: {j}, k: {k}, len: {content}")
```

```python
from tqdm import tqdm

cols = ['context', 'question', 'answer']
new_df_train = pd.DataFrame(columns=cols)

data = data_qas_train

for i in tqdm(range(len(data))):
    for j in range(len(data[i]['paragraphs'])):
        for k in range(len(data[i]['paragraphs'][j]['qas'])):
    
            new_df_train = new_df_train.append({
                'context': data[i]['paragraphs'][j]['context'],
                'question': data[i]['paragraphs'][j]['qas'][k]['question'],          
                'answer': {
                    
                    "text": data[i]['paragraphs'][j]['qas'][k]['answers'][0]['text'], 
                           
                    "answer_start": data[i]['paragraphs'][j]['qas'][k]['answers'][0]['answer_start'], 
                    
                    "answer_end": data[i]['paragraphs'][j]['qas'][k]['answers'][0]['answer_start'] +\
                                        len(data[i]['paragraphs'][j]['qas'][k]['answers'][0]['text'])}
                
            }, ignore_index=True)
```
