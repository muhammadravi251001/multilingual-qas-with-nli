```python
prefix = "https://huggingface.co/muhammadravi251001/fine-tuned-FilteringNLI-"
suffix = "/blob/main/results/evaluation/metric_result.txt"

suffix_df = "/raw/main/results/output/output_df.csv"
suffix_squadid_df = "/resolve/main/results/output/output_df.csv"
```

```python
def take_dataframe(url): 
    
    response = requests.get(url)
    
    if response.status_code == 200:
        csv_data = StringIO(response.text)
        df = pd.read_csv(csv_data, index_col=0)
        df = df[['Context', 'Question', 'Prediction Answer', 'Rec. Pred Answer', 'Gold Answer', 'Properties']]
        df.fillna('', inplace=True)
    
    else:
        print("Failed to download CSV")
    
    return df
```

```python
def analyse_mb_and_hb(df):
    
    benar = "Benar"
    
    r_qa_mb__r_nli_mb = 0
    r_qa_mb__w_nli_mb = 0
    w_qa_mb__r_nli_mb = 0
    w_qa_mb__w_nli_mb = 0
    
    r_qa_hb__r_nli_hb = 0
    r_qa_hb__w_nli_hb = 0
    w_qa_hb__r_nli_hb = 0
    w_qa_hb__w_nli_hb = 0
    
    for i in range(len(df)):
        
        qa_mb = df["QA Generated Answer MB"][i]
        qa_hb = df["QA Generated Answer HB"][i]
        nli_mb = df["NLI Validated Answer MB"][i]
        nli_hb = df["NLI Validated Answer HB"][i]
        
        # MB: metric-based section
        
        if benar in qa_mb and benar in nli_mb:
            r_qa_mb__r_nli_mb += 1
        elif benar in qa_mb and benar not in nli_mb:
            r_qa_mb__w_nli_mb += 1
        elif benar not in qa_mb and benar in nli_mb:
            w_qa_mb__r_nli_mb += 1
        elif benar not in qa_mb and benar not in nli_mb:
            w_qa_mb__w_nli_mb += 1
        
        # HB: human-based section
        
        if benar in qa_hb and benar in nli_hb:
            r_qa_hb__r_nli_hb += 1
        elif benar in qa_hb and benar not in nli_hb:
            r_qa_hb__w_nli_hb += 1
        elif benar not in qa_hb and benar in nli_hb:
            w_qa_hb__r_nli_hb += 1
        elif benar not in qa_hb and benar not in nli_hb:
            w_qa_hb__w_nli_hb += 1
            
    assert r_qa_mb__r_nli_mb + r_qa_mb__w_nli_mb + w_qa_mb__r_nli_mb + w_qa_mb__w_nli_mb == len(df)
    assert r_qa_hb__r_nli_hb + r_qa_hb__w_nli_hb + w_qa_hb__r_nli_hb + w_qa_hb__w_nli_hb == len(df)
    
    print("MB: metric-based section")
    print("Banyak QA Generated Answer benar dan NLI Validated Answer benar:", r_qa_mb__r_nli_mb)
    print("Banyak QA Generated Answer benar menjadi NLI Validated Answer salah:", r_qa_mb__w_nli_mb)
    print("Banyak QA Generated Answer salah menjadi NLI Validated Answer benar:", w_qa_mb__r_nli_mb)
    print("Banyak QA Generated Answer salah dan NLI Validated Answer salah:", w_qa_mb__w_nli_mb)
    print()
    
    print("HB: human-based section")
    print("Banyak QA Generated Answer benar dan NLI Validated Answer benar:", r_qa_hb__r_nli_hb)
    print("Banyak QA Generated Answer benar menjadi NLI Validated Answer salah:", r_qa_hb__w_nli_hb)
    print("Banyak QA Generated Answer salah menjadi NLI Validated Answer benar:", w_qa_hb__r_nli_hb)
    print("Banyak QA Generated Answer salah dan NLI Validated Answer salah:", w_qa_hb__w_nli_hb)
    print()
```
